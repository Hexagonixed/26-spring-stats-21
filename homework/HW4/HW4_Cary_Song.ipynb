### Part 6) Histograms of body mass by species

Create **separate** histograms of `body_mass_g` for each species (three panels).
    
- Use the same binning across panels so comparisons are fair.  
- Label axes clearly


```python
fig, axes = plt.subplots(1, 3, figsize = (16, 6), sharey = True, sharex = True)
fig.suptitle("Body Mass by Species")

axes[0].hist(penguins_clean.loc[penguins_clean["species"] == "Adelie", "body_mass_g"], bins = 15)
axes[0].set_title("Adelie")
axes[0].set_xlabel("Body mass (g)")
axes[0].set_ylabel("Count")

axes[1].hist(penguins_clean.loc[penguins_clean["species"] == "Chinstrap", "body_mass_g"], bins = 15)
axes[1].set_title("Chinstrap")
axes[1].set_xlabel("Body mass (g)")
axes[1].set_ylabel("Count")

axes[2].hist(penguins_clean.loc[penguins_clean["species"] == "Gentoo", "body_mass_g"], bins = 15)
axes[2].set_title("Gentoo")
axes[2].set_xlabel("Body mass (g)")
axes[2].set_ylabel("Count")
```

### Part 8) Scatter: bill length vs bill depth

### Part 9) Stacked bars: sex proportions by species

Compute counts of `sex` within each `species`, convert to **proportions**, and make a **stacked bar chart**.  

- Bars should be species on the x‑axis with male/female proportions stacked to 1.0. 

```python
pd.crosstab(penguins_clean["species"], penguins_clean["sex"], normalize = 'index') \
    .plot(kind = "bar", stacked = True)

plt.title("Porportion of Sex by Species")
```
